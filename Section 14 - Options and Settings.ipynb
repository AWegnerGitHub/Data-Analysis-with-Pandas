# Section 14; Part 170
Changes pandas options with attributes and dot syntax

```python
# Generate large fake dataset
#   Create random integers between 0 and 100 and put it in a 1000x50 array
data = np.random.randint(0, 100, [1000, 50])
data
```

```python
# Create a dataframe from the numpy array
df = pd.DataFrame(data)
df.head()
```

```python
# Options
#  - max_rows -> Number of rows shown
#  - max_columns -> Number of columns shown
print pd.options.display.max_rows
print pd.options.display.max_columns
```

```python
# Set new values
pd.options.display.max_rows = 4
df
```

```python
# Reset to 60
pd.options.display.max_rows = 60
```

# Section 14; Part 171
Changing options with methods

```python
df
```

  - `get_option` - call on pandas library, pass it the option name
  - `set_option` - first parameter is option name, second is value
  - `reset_option` - Reset to default
  - `describe_option` - Explain an option and provides the current and default values

```python
pd.get_option("max_rows")
```

```python
pd.set_option("max_rows", 4)
df
```

```python
pd.reset_option('max_rows')
df
```

```python
pd.describe_option('max_rows')
```

# Section 14; Part 172
The `precision` option

 - Alter the number of digits that appears (without changing the actual number) after a decimal place

```python
# randn -> Dataset of statistical diviations
df = pd.DataFrame(np.random.randn(5,5))
df
```

```python
# Only show 2 decimal points, without changing underlying data
pd.get_option('precision')
```

```python
pd.set_option('precision', 2)
df
```

```python
# If reset, the original numbers remain
pd.reset_option('precision')
df
```
